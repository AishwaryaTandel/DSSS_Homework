```python
import numpy as np
from numba import jit

@jit(nopython=True)
def approximate_pi_numba(n):
    pi_2 = 1
    nom, den = 2.0, 1.0
    for i in range(n):
        pi_2 *= nom / den
        if i % 2:
            nom += 2
        else:
            den += 2
    return 2 * pi_2

# Test the Numba-optimized function
result_numba = approximate_pi_numba(1000000)
print(result_numba)
```
